## This loads the census-data and aggregates the data to the first 3 digits of the zip-code to fit with Lending Club data

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/census_zipcode_level.csv")
```

```python
df.head()
```

```python
df['Zip'] = df['Zip'].astype(str).str.slice(0,3)
df.head()
```

```python
df2 = df.groupby('Zip').sum()
```
