This notebook establishes a baseline model by setting all loans to perform according to the most common value in the loan_status, which is fully paid. 

A model predicting all loans to perform and be fully paid off, will achieve an accuracy of 0.76357, and will work as our baseline model.

```python
import numpy as np
import pandas as pd
```

```python
# There are a lot of columns - increase display
pd.set_option('display.max_columns', None)
pd.options.display.max_rows = 150
```

```python
# read in the 2016-2017 data set
df = pd.read_csv('/Users/oeyvind/Documents/Studier/CS109A/Project/LendingClubFiles/data_cleaned_2016_2017.csv', low_memory = False)
```

```python
#Check that the data looks ok
df.head()
```

```python
#Ensure that the data only contain the two alternatives for finished loans, charged off and fully paid
df.loan_status.unique()
```

```python
#Find the percentage of loans that ends up as fully paid
df.loan_status[df.loan_status == 'Fully Paid'].count() / (df.loan_status[df.loan_status == 'Charged Off'].count()+df.loan_status[df.loan_status == 'Fully Paid'].count())
```

```python
#Find the percentage of loans that ends up as fully paid
baseline_accuracy = np.sum(df.loan_status == 'Fully Paid') / len(df.loan_status)

print("The accuracy of a model predicting all loans to be fully paid is {}".format(baseline_accuracy))
```
